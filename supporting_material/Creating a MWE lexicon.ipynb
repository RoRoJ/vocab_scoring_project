# Creating a MWE lexicon
This notebook shows how the json file of MWEs from wiktionary (download available at: http://www.cs.cmu.edu/~ark/LexSem/mwelex-1.0.zip) was converted to a list of MWEs and saved with pickle as a list object

```python
import json
import pickle
```

```python
#https://stackoverflow.com/questions/19201290/how-to-save-a-dictionary-to-a-file
def save_obj(obj, name ):
    with open(r'C:/Users/rowena/Documents/MSC/Project/obj/'+ name + '.pkl', 'wb') as f:
        pickle.dump(obj, f, pickle.HIGHEST_PROTOCOL)
```

```python
json_data = []
for line in open(r'C:/Users/rowena/Documents/MSC/Project/PVs_Exprs/enwikt.json', "r"):
    json_data.append(json.loads(line))
mwes_list = []
for line in json_data:
    mwes_list.append(dict(line)['words'])
mwes_list = sorted(mwes_list,key = len)
```

```python
mwes_list
```

```python
save_obj(mwes_list, 'mwes_list')
```
